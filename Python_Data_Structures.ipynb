# Data Structures

### Lists

```python
# Declaration of a List

# Empty List
empty_list=[]

# List of elements
list1 = ['hello', 'world', [1, 2, 1+2]]  # Multiple datatypes can be stored in a list.
print("List:\t\t", list1)

# Appending an element to a list
list2 = ['bye']
list1.append(list2)
print("\nAppending:\t", list1)

# Extending the list
list1.extend(list2)
print("\nExtending:\t", list1)

list1 = list1 + list2
print("\nExtending:\t", list1)

# Inserting an item in a list
list1.insert(2, "HI")
print("\nInserting:\t", list1)

# Removing an item from a list
list1.remove(['bye']) # 'remove' only removes first occurance of the element. Here, only one element is ['bye']
print("\nRemoving:\t", list1)

# Reversing the list
list1.reverse()
print("\nReversing:\t", list1)

# Length of list
print("\nLength of list:\t", len(list1))

# Count of an element in a list
print("\nCount:\t\t", list1.count('bye'))

# Slicing of a list
print("\nSliced list:\t", list1[0:6:2])     # Start : End : Step size
```

### Tuples

```python
# Declaration of a Tuple

# Empty Tuple
empty_tuple = ()

# Tuple of elements
tuple1 = ('hello', 'world', [1, 2 , 1+2])  # Immutable data structure
print("Tuple:\t\t\t", tuple1)

# Changing an item (list) in a tuple which is mutable.
tuple1[2][2] = tuple1[2][2] + 3 
print("\nChanging:\t\t", tuple1)

# Concatenation
tuple2 = (1, 20, 3) + (4, 50, 6)
print("\nConcatenation:\t\t", tuple2)

# Sorting
new_tuple2 = sorted(tuple2)
print("\nSorted tuple:\t\t", tuple2)

# Max/min value in tuple
print("\nMax:\t\t\t", max(tuple2))
print("\nMin:\t\t\t", min(tuple2))

# Sum of all numeric values in tuple
print("\nSum:\t\t\t", sum(tuple2))

# Repitition
tuple1 = tuple1 * 4
print("\nRepitition:\t\t", tuple1)

# Deletion
del tuple2

# Count of an element
print("\nCount of 'hello':\t", tuple1.count('hello'))

# Length of tuple
print("\nLength of tuple:\t", len(tuple1))
```

### Dictionaries

```python
# Declaration of a Dictionary
dict1 = {'a':'hello', 'b': 'world', 'c': [1, 2, 1+2]}  # Multiple datatypes can be stored in a dictionary.
print("Dictionary:\t\t", dict1)

# Printing the value related to key 'c'
print("\nValue of key 'c':\t", dict1.get('c'))

# Adding a key-value pair
dict1['d'] = 'bye'
print("\nAddition:\t\t", dict1)

# Deleting a key-value pair
dict1.pop('c')
print("\nDeletion:\t\t", dict1)

print("\nAll methods which can be applied to dictionary:\n", dir(dict1))

# Making a copy of a dictionary
dict2 = { k + '2' : v + '2' for k,v in dict1.items() }
print("\nCopied Dictionary:\t", dict2)

# Printing key-value pairs
print("\nKey, Value Pairs:\t", dict1.items())

# Printing only keys
print("\nKeys:\t\t\t", dict1.keys())

# Printing key-value pairs
print("\nValues:\t\t\t", dict1.values())
```

### Strings

```python
# Declaration of a String
str1 = 'Hello, world!'
print("Last character of string 1:\t", str1[12])

str2 = "Hola!"
print("\nLast character of string 2:\t", str2[-1])

str3 = '''Namaste!'''
print("\nFew characters of string 3:\t", str3[1:6])

# Concatenation
print("\nConcatenation:\t\t\t", str1 + str2)
print("\nConcatenation:\t\t\t", str3 * 3)

# String Membership Test
print("\nString Membership Test:\t\t", 'l' in str3, "\t\t\t(Checking if character 'l' present in String 3)")

# Upper case characters
print("\nUpper Case:\t\t\t", str3.upper())

# Splitting of a string
str4 = str1.split()
print("\nSpliting of string 1:\t\t", str4)

# Joining of split words
print("\nJoining of split words:\t\t", '-'.join(str4))

# Reversing the string character by character
print("\nReversed string:\t\t", list(reversed(str2)))

# Replacing a part of a string with a substring
print("\nReplaced word in string:\t", str1.replace('Hello', 'Bye'))
```
